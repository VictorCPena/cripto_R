Importando Bibliotecas

```r
library(rvest)
library(dplyr)
library(stringr)
library(ggplot2)
```

Engenharia de Dados

```r
url <- "https://br.investing.com/crypto/currencies"

page <- read_html(url)
```

```r
nome_cripto <- page %>% 
  html_nodes("[class='crypto-coins-table_cellNameText__aaXmK']") %>%
  html_text()
nome_cripto
```

```r
symbol_cripto <- page %>% 
  html_nodes("[class='text-common-grey pt-0.5 text-xs']") %>%
  html_text()
symbol_cripto
```

```r
preco_cripto <- page %>% 
  html_nodes("[class='datatable_cell__LJp3C datatable_cell--align-end__qgxDQ text-secondary !text-sm crypto-coins-table_thirdMobileCell__f8EsE']") %>% # nolint
  html_text()
preco_cripto
```

```r
var_cripto<- page %>% 
  html_nodes(".datatable_cell--down___c4Fq , .datatable_cell--up__hIuZF") %>% # nolint
  html_text()
var_cripto
```

```r
capitalizacao_cripto <- page %>% 
  html_nodes("[class='datatable_cell__LJp3C datatable_cell--align-end__qgxDQ !text-sm']") %>% # nolint
  html_text()
capitalizacao_cripto
```

```r
volume24h_cripto <- page %>% 
  html_nodes("[data-test='volume-cell']") %>% # nolint
  html_text()
volume24h_cripto
```

```r
volumetotal_cripto <- NULL
cripto <- NULL

for (i in seq(1, length(capitalizacao_cripto), by = 3)) {
  if (i + 2 <= length(capitalizacao_cripto)) {
    primeiro_elemento <- capitalizacao_cripto[i]
    terceiro_elemento <- capitalizacao_cripto[i + 2]
    cripto <- c(cripto, primeiro_elemento)
    volumetotal_cripto <- c(volumetotal_cripto, terceiro_elemento)
  }
}

capitalizacao_cripto <- cripto

print(capitalizacao_cripto)
print(volumetotal_cripto)
```

```r
capitalizacao_cripto <- gsub(",", ".", capitalizacao_cripto)
capitalizacao_cripto <- gsub("\\$", "", capitalizacao_cripto)

volume24h_cripto <- gsub(",", ".", volume24h_cripto)
volume24h_cripto <- gsub("\\$", "", volume24h_cripto)

preco_cripto <- gsub("\\$", "", preco_cripto)
var_cripto <- gsub(",", ".", var_cripto)

var_cripto <- as.numeric(gsub("[+,]|%", "", var_cripto))
volumetotal_cripto <- as.numeric(gsub("%", "", volumetotal_cripto))
```

```r
var24h_cripto <- var_cripto[seq(1, length(var_cripto), by = 2)]
var7d_cripto <- var_cripto[seq(2, length(var_cripto), by = 2)]
```

```r
converter_para_numerico <- function(valor) {
  if (grepl("B", valor)) {
    valor_sem_sufixo <- gsub("B", "", valor)
    return(as.numeric(valor_sem_sufixo) * 1e9)
  } else if (grepl("M", valor)) {
    valor_sem_sufixo <- gsub("M", "", valor)
    return(as.numeric(valor_sem_sufixo) * 1e6)
  } else if (grepl("K", valor)) {
    valor_sem_sufixo <- gsub("K", "", valor)
    return(as.numeric(valor_sem_sufixo) * 1e3)
  } else {
    return(as.numeric(valor))
  }
}

capitalizacao_cripto <- sapply(capitalizacao_cripto, converter_para_numerico)
volume24h_cripto <- sapply(volume24h_cripto, converter_para_numerico)
```

```r
converter_preco <- function(valor) {
  valor <- gsub(",", ".", valor)

  partes <- strsplit(valor, "\\.")[[1]]

  if (length(partes) > 1) {
    parte_inteira <- paste(partes[-length(partes)], collapse = "")
    parte_decimal <- partes[length(partes)]
  } else {
    parte_inteira <- partes[1]
    parte_decimal <- "0"
  }

  num_zeros <- nchar(parte_decimal)
  divisor <- 10 ^ num_zeros 
  resultado <- as.numeric(parte_inteira) + as.numeric(parte_decimal) / divisor
  return(resultado)
}

preco_cripto <- sapply(preco_cripto, converter_preco)
```

```r
suppressWarnings({
  dados_cripto <- data.frame(
    nome_cripto,
    symbol_cripto,
    preco_cripto,
    var24h_cripto,
    var7d_cripto,
    capitalizacao_cripto,
    volume24h_cripto,
    volumetotal_cripto
  )
})
```

```r
dados_plot <- dados_cripto[, c("symbol_cripto", "var24h_cripto", "var7d_cripto")] #nolint
dados_plot <- transform(dados_plot, total_var24h_7d = var24h_cripto + var7d_cripto) #nolint

grafico <- ggplot(dados_plot, aes(x = symbol_cripto, y = total_var24h_7d, fill = factor(symbol_cripto))) + #nolint
  geom_bar(stat = "identity") +
  geom_errorbar(aes(ymin = -5, ymax = 5), width = 0.2) +
  labs(x = "Criptomoeda", y = "Variação de Preço", fill = "Variação") +
  ggtitle("Variação de Preço nas Últimas 24 Horas e 7 Dias") +
  theme_minimal() +
  ylim(-5, 5)


media_var24h <- mean(dados_plot$var24h_cripto)
media_var7d <- mean(dados_plot$var7d_cripto)

grafico +
  geom_hline(yintercept = media_var24h, linetype = "dashed", color = "blue") +
  geom_hline(yintercept = media_var7d, linetype = "dashed", color = "red")
```

```r
dados_plot <- dados_cripto %>%
  mutate(diff_var = var7d_cripto - var24h_cripto,
         var_type = ifelse(diff_var >= 0, "Valorização", "Desvalorização")) %>%
  arrange(desc(diff_var))

top_valorizadas <- head(dados_plot, 10)
top_desvalorizadas <- tail(dados_plot, 10)

ggplot(top_valorizadas, aes(x = reorder(symbol_cripto, diff_var), y = diff_var, fill = var_type)) + #nolint
  geom_bar(stat = "identity") +
  scale_fill_manual(values = c("Valorização" = "green", "Desvalorização" = "red")) + #nolint
  labs(x = "Criptomoeda", y = "Diferença de Variação (7d - 24h)", fill = "Variação") + #nolint
  ggtitle("Top 10 Criptomoedas com Maior Valorização entre 7d e 24h") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))

ggplot(top_desvalorizadas, aes(x = reorder(symbol_cripto, diff_var), y = diff_var, fill = var_type)) + #nolint
  geom_bar(stat = "identity") +
  scale_fill_manual(values = c("Valorização" = "green", "Desvalorização" = "red")) + #nolint
  labs(x = "Criptomoeda", y = "Diferença de Variação (7d - 24h)", fill = "Variação") + #nolint
  ggtitle("Top 10 Criptomoedas com Maior Desvalorização entre 7d e 24h") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
```
